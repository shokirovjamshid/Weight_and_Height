### `weight-height.csv` faylini yuklab oling.

 **Gender** - ustunini listga 0 va 1 ko'rinishda joylashtiring

 **Weight** - ustunini Pound da berilgan Kg ko'rinishda joylashtiring (1 kg = 2.205 pound).

 **Height** - ustunini inch da berilgan sm ko'rinishda listga joylashtiring (2.54 sm = 1 inch).

```python
# import library
import csv
import matplotlib.pyplot as plt
```

```python
def read_csv_data(file_path):
    """
    Read csv file
    Args:
        file_path(str): file path
    Returns:
        list: data split row
    """
    #WRITE YOUR CODE HERE
    f = open(file_path, mode='r+')
    data = csv.reader(f,delimiter=',')
    return list(data)[:-1]
```

```python
def get_data(data):
    """
    Get data from list.
    Gender: Change Male to 0 and Female to 1
    Weight: Place the column in the Kg view given in Pound (1 kg = 2,205 pound).
    Height: Place the column on the list in the cm view given in inches (2.54 cm = 1 inch).
    Args:
        data(list): data split row
    Returns:
        tuple: gender, weight and height

    """
    #WRITE YOUR CODE HERE
    gender = []
    weight = []
    height = []
    for i in data[1:]:
      weight.append(float(i[2])/2.205)
      height.append(float(i[1])*2.54)
      if i[0]=='Male':
        gender.append(0)
      elif i[0]=='Female':
        gender.append(1)
    return gender,weight,height

# print(get_data(read_csv_data('weight-height.csv')))

```

# Scatter plot
**-** **Weight** va **height** ma'lumotlarni **gender** orqali ajratib alohida listlarga joylashtiring.

**-** Ma'lumotlarni tasvirlang.

```python
plt.figure(dpi=150)

def show(gender, weight, height):
    """
    Show data.
    Args:
        gender(list): 0 and 1
        weight(list): weight
        height(list): height
    Returns:
        None
    """
    w_male = []
    w_female = []
    h_male = []
    h_female = []
    #WRITE YOUR CODE HERE
    for g,w,h in zip(gender,weight,height):
      if g==0:
        w_male.append(w)
        h_male.append(h)
      elif g==1:
        w_female.append(w)
        h_female.append(h)
    plt.scatter(h_female,w_female,color='r',label='Female',alpha=0.65)
    plt.scatter(h_male,w_male,color='black',label='Male',alpha=0.65)
    plt.legend(loc=1,ncol=2,bbox_to_anchor=(0.41,1.1))
    plt.show()


data = read_csv_data('weight-height.csv')
gender, weight, height = get_data(data)

show(gender, weight, height)
```
